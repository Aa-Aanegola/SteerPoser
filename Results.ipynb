```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Execution Rate', 'Preference Alignment']
voxposer = [90, 28]
steerposer = [80, 65]

x = np.arange(len(labels))  # label locations
width = 0.35  # width of the bars

# Create figure and axes
fig, ax = plt.subplots(figsize=(6, 4))

# Plot bars
rects1 = ax.bar(x - width/2, voxposer, width, label='Voxposer', color='#4C72B0')
rects2 = ax.bar(x + width/2, steerposer, width, label='SteerPoser', color='#55A868')

# Add labels, title, legend, etc.
ax.set_ylabel('Score (%)', fontsize=12)
ax.set_title('Comparison of Voxposer and SteerPoser', fontsize=14, pad=10)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11)
ax.set_ylim(0, 100)
ax.legend(fontsize=11)

# Add value labels on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

# Improve layout
fig.tight_layout()
plt.show()
```
